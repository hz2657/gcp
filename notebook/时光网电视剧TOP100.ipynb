# 1. scrape data by beautiful soup 

```python
#------scrape TV show info----------------------------------
import requests, bs4

headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}

for x in range(1,11):
    if x == 1: 
        url = 'http://www.mtime.com/top/tv/top100/'
    else: 
        url = 'http://www.mtime.com/top/tv/top100/index-' + str(x) + '.html'
    res = requests.get(url, headers=headers)
    bs = bs4.BeautifulSoup(res.text, 'html.parser')
    bs = bs.find(class_="top_list") 
    lis = bs.find_all('li')
    for li in lis:
        num = li.find(class_='number').text
        title = li.find(class_='px14 pb6').text
        if li.find(class_="mt3") != None:
            comment = li.find(class_="mt3").text
            print('——' + '\n' + num + title + '\n' + '推荐语：' + comment)
        else:
            print('——' + '\n' + num + title)
        
        
```

# 2. save data in Excel file

```python
import openpyxl

wb=openpyxl.Workbook()  
sheet=wb.active 
sheet.title='TV shows' 

sheet['A1'] ='No.'     #加表头，给A1单元格赋值
sheet['B1'] ='title'   #加表头，给B1单元格赋值

headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}

for x in range(1,11):
    if x == 1: 
        url = 'http://www.mtime.com/top/tv/top100/'
    else: 
        url = 'http://www.mtime.com/top/tv/top100/index-' + str(x) + '.html'
    res = requests.get(url, headers=headers)
    bs = bs4.BeautifulSoup(res.text, 'html.parser')
    bs = bs.find(class_="top_list") 
    lis = bs.find_all('li')
    for li in lis:
        num = li.find(class_='number').text
        title = li.find(class_='px14 pb6').text
        if li.find(class_="mt3") != None:
            comment = li.find(class_="mt3").text
            print('——' + '\n' + num + title + '\n' + '推荐语：' + comment)
        else:
            print('——' + '\n' + num + title)
        sheet.append([num,title])    
         
wb.save('Top 100 TV shows.xlsx')            
            
```
